Q1.  Write a program to find a number is odd or even

```python
def check(num):
  if(num%2==0):
    return "Even"
  else:
    return "Odd"

print(check(10))
print(check(5))
```

Q2.  Write a program to find if a number is prime or not

```python
def check(num):
  check=0;
  for i in range(2,num-1):
    if(num%i==0):
      check=check+1
  if(check!=0):
    print("Not Prime")
  else:
    print("Prime")

check(5)
check(10)
```

Q3. Write a program to create triangle/pyramid pattern

```python
n = 5

for i in range(1, n+1):
    for j in range(i):
        print("*", end=" ")
    print()
```

Q4. Write a program to find the factorial of a number.

```python
def factorial(num):
  if num==0:
    return 1
  else:
    return num*factorial(num-1)
factorial(4)
```

Q7. Write a program to design a calculator.

```python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    return x / y

print("Choose Operation.")
print("1.Add")
print("2.Subtract")
print("3.Multiply")
print("4.Divide")

choice = input("Enter choice (1/2/3/4): ")

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

if choice == '1':
    print(num1, "+", num2, "=", add(num1,num2))

elif choice == '2':
    print(num1, "-", num2, "=", subtract(num1,num2))

elif choice == '3':
    print(num1, "*", num2, "=", multiply(num1,num2))

elif choice == '4':
    print(num1, "/", num2, "=", divide(num1,num2))

else:
    print("Invalid input")
```

Q9. Write a program to create fibonnaci series upto n number.

```python
n = int(input("Enter the value of n: "))
fib = [0, 1]

for i in range(2, n):
    next_fib = fib[i-1] + fib[i-2]
    fib.append(next_fib)

print("The first", n, "Fibonacci numbers are:")
for num in fib:
    print(num)
```

Q10. Write a python program to generate first n Prime number.

```python
n = int(input("Enter the value of n: "))
primes = []
i = 2
while len(primes) < n:
    is_prime = True
    for j in range(2, int(i ** 0.5) + 1):
        if i % j == 0:
            is_prime = False
            break
    if is_prime:
        primes.append(i)
    i += 1

print("The first", n, "prime numbers are:")
for num in primes:
    print(num)
```
